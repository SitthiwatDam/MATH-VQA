```python
from transformers import Pix2StructProcessor, Pix2StructForConditionalGeneration, AutoModelForCausalLM, AutoTokenizer
# from conversation import get_default_conv_template
import torch
from PIL import Image
import matplotlib.pyplot as plt

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
```

```python
# chartsum_model_ids = 'google/matcha-chart2text-statista'

processor = Pix2StructProcessor.from_pretrained('google/matcha-chart2text-pew')
model = Pix2StructForConditionalGeneration.from_pretrained("./model/chart2text-autochart-alt").to(device)
```

```python
image = Image.open("./content/images/00035547003867.png").convert("RGB")
display(image)
```

```python
inputs = processor(images=image, text="What is this chart about?", return_tensors="pt", add_special_tokens=True, max_patches=1024).to(device)
predictions = model.generate(**inputs, max_new_tokens=512)
chart_sum = processor.decode(predictions[0], skip_special_tokens=True)

print(f"Chart summary:\n{chart_sum}")
```
