# Matcha-chart2text-pew

Note
- Visual Encoder. Trained on pew dataset
- Have similar model but trained with statista dataset
- The very first 100 dataset of chartqa is more similar to pew dataset. It is possible that some of the latter part is statista (need to see into it)

```python
from transformers import Pix2StructProcessor, Pix2StructForConditionalGeneration
import requests
from PIL import Image
```

```python
processor = Pix2StructProcessor.from_pretrained('google/matcha-chart2text-pew')
model = Pix2StructForConditionalGeneration.from_pretrained('google/matcha-chart2text-pew').to(0)
```

```python
url = "https://github.com/vis-nlp/ChartQA/blob/main/ChartQA%20Dataset/train/png/00035547003867.png?raw=true"
image = Image.open(requests.get(url, stream=True).raw)
```

```python
display(image)
```

```python
inputs = processor(images=image, text="What is this chart about?", return_tensors="pt").to(0)
predictions = model.generate(**inputs, max_new_tokens=512)
print(processor.decode(predictions[0], skip_special_tokens=True))
```

```python
prompt = "Please generate a descriptive analysis of the quantitative data presented in the chart below. \
    Include key statistics, trends, and insights."

inputs = processor(images=image, text=prompt, return_tensors="pt").to(0)
predictions = model.generate(**inputs, max_new_tokens=512)
print(processor.decode(predictions[0], skip_special_tokens=True))
```

```python
prompt = "Translate the numerical information depicted in the chart into a comprehensive textual summary. \
    Describe the main patterns, variations, and relationships evident in the data."

inputs = processor(images=image, text=prompt, return_tensors="pt").to(0)
predictions = model.generate(**inputs, max_new_tokens=512)
print(processor.decode(predictions[0], skip_special_tokens=True))
```
